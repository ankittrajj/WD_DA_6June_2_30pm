```python
# pie chart-----> percentage distribution----->
# no axis
# comparision

import matplotlib.pyplot as plt
import numpy as np
```

```python
sales = [45,56,23,89]
mylabels = ['India',"USA","Pakistan",'Srilanka']
plt.pie(sales,labels=mylabels)
```

```python
sales = [45,56,23,89]
mylabels = ['India',"USA","Pakistan",'Srilanka']
plt.pie(sales,labels=mylabels,autopct = '%.1f')
```

```python
# subplot------> to see multiple graphs...
x = [34,45,67,89]
y = [90,78,34,12]
plt.subplot(2,2,1)
plt.plot(x,y)

x = [1,2,3,4]
y = [90,78,34,12]
plt.subplot(1,2,2)
plt.plot(x,y)
```

```python
# scatter plot----->

x = np.array([2,6,10,14])
y = np.array([0,1,2,3])
plt.scatter(x,y)
plt.plot(x,y)
```
